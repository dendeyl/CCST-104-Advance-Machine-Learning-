# **1. Data Cleaning and Preparation**

**• Task:**

Clean the dataset by handling missing values, removing duplicates, and converting fields like 'Reviews', 'Size', 'Installs', and 'Price' to a numerical format for analysis.

**• Skills Assessed:**

```python
data = pd.read_csv('/content/googleplaystore.csv')

data.tail()
```

```python
def convert_to_numeric(value):

    num, sign = value[:-1], value[-1]

    try:
      if sign == '+':
        return 1000

      if sign.lower() == 'm':
        return float(num)*1000000
      if sign.lower() == 'k':
        return float(num)*1000

    except:
      return 0



data['Size'] = [convert_to_numeric(value) for value in data['Size']]
data['Size'].fillna(data['Size'].mean(),inplace=True)
```

```python
def convert_review(value):
  try:
    return int(value)
  except:
    num, sign = value[:-1], value[-1]
    return float(num)*1000000


data['Reviews'] = [convert_review(i) for i in data['Reviews']]
```

```python
data.isna().sum()
```

```python
def convert_installs(category):
    if category == 'Free':
        return 0
    elif '+' in category:
        return int(category.replace('+', '').replace(',', ''))
    else:
        return int(category)


data['Installs'] = [convert_installs(category) for category in data['Installs']]

# data
```

```python
data['Price'].unique()
```

```python
def convert_prices(price):
  if price == 'Everyone':
   return 0
  else:
    numeric_price = float(price.replace('$', '').replace(',', ''))
    return numeric_price

data['Price'] = [convert_prices(prices) for prices in data['Price']]
```
